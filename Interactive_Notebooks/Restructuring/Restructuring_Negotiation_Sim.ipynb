# Restructuring Negotiation Simulation

## Role: Senior Lender
You are a Senior Lender to **DistressedCo**, a manufacturing firm facing liquidity issues. The company has breached its leverage covenants and missed a principal payment.

**Objective:** Maximize recovery value while deciding whether to force a liquidation or agree to a restructuring proposal.

```python
import ipywidgets as widgets
from IPython.display import display

print("Situation Update: The company proposes a debt-for-equity swap. They want to convert 50% of your debt into 80% equity ownership.")

decision = widgets.Dropdown(
    options=['Accept Proposal', 'Reject & Counter (70% Debt, 90% Equity)', 'Force Liquidation'],
    value='Accept Proposal',
    description='Decision:',
)

def on_change(change):
    if change['new'] == 'Force Liquidation':
        print("Outcome: Liquidation value is only 40 cents on the dollar. You take a significant loss.")
    elif change['new'] == 'Accept Proposal':
        print("Outcome: You now control the company. Upside is possible if operations turn around.")
    else:
        print("Outcome: Negotiation continues. The Private Equity sponsor threatens to walk away.")

decision.observe(on_change, names='value')
display(decision)
```
